```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stats = open("stats.txt", 'r')
data = list(map(lambda elem: elem.split(), stats.readlines()))
n = int(data[0][1])
m = int(data[0][2])
tests = len(data) - 1
ps = [int(data[i][3]) for i in range(tests)]
times = [float(data[i][0][:-1]) for i in range(tests)]

qsort_stats = open("qsort_stats.txt", 'r')
qsort_data = list(map(lambda elem: elem.split(), qsort_stats.readlines()))
qtimes = [float(qsort_data[i][0][:-1]) for i in range(tests)]

sp = np.ones(tests) * float(data[-1][0][:-1]) / np.array(times) 
ep = sp / np.array(ps)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ps, times, label="T(P)")
plt.plot(ps, qtimes, label="QuickSort T(P)")
plt.plot(ps, sp, label="S(P)")
plt.plot(ps, ep, label="E(P)", color='black')
plt.legend()
plt.show()
```
